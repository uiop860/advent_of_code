```python
all_lines: list[str] = []
with open("./data.txt") as file:
    for line in file:
        all_lines.append(list(line.strip()))


def count_xmas_occurrences(grid, word="XMAS"):
    rows, cols = len(grid), len(grid[0])
    word_length = len(word)
    directions = [
        (0, 1),  # right
        (0, -1),  # left
        (1, 0),  # down
        (-1, 0),  # up
        (1, 1),  # diagonal down-right
        (1, -1),  # diagonal down-left
        (-1, 1),  # diagonal up-right
        (-1, -1),  # diagonal up-left
    ]

    def is_valid(x, y):
        return 0 <= x < rows and 0 <= y < cols

    def matches(x, y, dx, dy):
        for i in range(word_length):
            nx, ny = x + i * dx, y + i * dy
            if not is_valid(nx, ny) or grid[nx][ny] != word[i]:
                return False
        return True

    count = 0
    for r in range(rows):
        for c in range(cols):
            for dx, dy in directions:
                if matches(r, c, dx, dy):
                    count += 1
    return count


# Example grid
grid = [
    "MMMSXXMASM",
    "MSAMXMSMSA",
    "AMXSXMAAMM",
    "MSAMASMSMX",
    "XMASAMXAMM",
    "XXAMMXXAMA",
    "SMSMSASXSS",
    "SAXAMASAAA",
    "MAMMMXMMMM",
    "MXMXAXMASX",
]

# Convert to a list of lists for easier indexing
# grid = [list(row) for row in grid]
print("Total XMAS occurrences:", count_xmas_occurrences(all_lines))
```

```python
all_lines: list[str] = []
with open("./data.txt") as file:
    for line in file:
        all_lines.append(list(line.strip()))


def count_xmas_patterns(grid):
    rows, cols = len(grid), len(grid[0])
    pattern_count = 0

    def is_valid(x, y):
        return 0 <= x < rows and 0 <= y < cols

    def check_xmas(r, c):
        # Check the "X" structure around center grid[r][c]
        if not is_valid(r - 1, c - 1) or not is_valid(r - 1, c + 1):
            return False
        if not is_valid(r + 1, c - 1) or not is_valid(r + 1, c + 1):
            return False

        # Diagonal elements
        top_left = grid[r - 1][c - 1]
        top_right = grid[r - 1][c + 1]
        center = grid[r][c]
        bottom_left = grid[r + 1][c - 1]
        bottom_right = grid[r + 1][c + 1]

        # Debug: Print cells being checked
        # print(f"Checking X-MAS at center ({r},{c}):")
        # print(f"  {top_left} . {top_right}")
        # print(f"  . {center} .")
        # print(f"  {bottom_left} . {bottom_right}")

        # Check all valid diagonal combinations
        return top_left + center + bottom_right in {
            "MAS",
            "SAM",
        } and top_right + center + bottom_left in {"MAS", "SAM"}

    for r in range(1, rows - 1):
        for c in range(1, cols - 1):
            if grid[r][c] == "A" and check_xmas(r, c):
                pattern_count += 1

    return pattern_count


# Example grid
grid = [
    "MMMSXXMASM",
    "MSAMXMSMSA",
    "AMXSXMAAMM",
    "MSAMASMSMX",
    "XMASAMXAMM",
    "XXAMMXXAMA",
    "SMSMSASXSS",
    "SAXAMASAAA",
    "MAMMMXMMMM",
    "MXMXAXMASX",
]

# Convert to a list of lists for easier indexing
grid = [list(row) for row in grid]
print("Total X-MAS patterns:", count_xmas_patterns(all_lines))
```
