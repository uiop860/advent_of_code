# Day 1: Rock Paper Scissors 

## Part 1

```python
with open('./input_data/day_2_input.txt') as f:
  data = f.read()
  

translate = {
  'A X' : '4',
  'B X' : '1',
  'C X' : '7',
  'A Y' : '8',
  'B Y' : '5',
  'C Y' : '2',
  'A Z' : '3',
  'B Z' : '9',
  'C Z' : '6'
}

for i in translate.keys():
  data = data.replace(i,translate.get(i))

result = sum([int(i) for i in data.split('\n')])
print(result)
```
